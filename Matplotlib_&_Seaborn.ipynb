# **MATPLOTLIB**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.7, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.colorbar(label='Color Intensity')
plt.show()
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 35]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# Histogram
np.random.seed(0)
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='lightcoral', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Pie Chart
labels = ['Part 1', 'Part 2', 'Part 3', 'Part 4']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

# **SEABORN**

```python
import seaborn as sns
```

```python
# Distribution Plot (Histogram with KDE)
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, color='salmon')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution Plot (Histogram with KDE)')
plt.show()
```

```python
min_length = min(len(x_scatter), len(y_scatter), len(colors))
df_scatter = pd.DataFrame({
    'X': x_scatter[:min_length],
    'Y': y_scatter[:min_length],
    'Color': colors[:min_length]
})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', hue='Color', size=sizes[:min_length], alpha=0.7, data=df_scatter, palette='viridis')
plt.title('Seaborn Scatter Plot')
plt.show()
```

```python
# Bar Plot with Seaborn
df_bar = pd.DataFrame({'Categories': categories, 'Values': values})
plt.figure(figsize=(8, 6))
sns.barplot(x='Categories', y='Values', data=df_bar, palette='mako')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Seaborn Bar Plot')
plt.show()
```

```python
# Box Plot
np.random.seed(0)
data_box = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_box, palette='Set2')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=data_box, palette='Set3')
plt.ylabel('Value')
plt.title('Violin Plot')
plt.show()
```

```python
# Pair Plot (for multiple numerical features)
iris = sns.load_dataset('iris')
plt.figure(figsize=(10, 10))
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
correlation_matrix = iris.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

```python
# Count Plot (for categorical data)
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris, palette='viridis')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count Plot of Iris Species')
plt.show()
```
